```python
#initialize environment, load modules/libraries

import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import umap
```

```python
loaded  = np.load('../data/shapes.npz')
images = loaded['images']
labels = loaded['labels'][:,0].astype(int)
```

```python
sample = 0.05
sample = int(sample*len(images))

images = images[:sample].reshape((sample, -1))/255
labels = labels[:sample]
```

```python
umap_model = umap.UMAP(n_neighbors=20, n_components=2, random_state=1711)
umap_fit = umap_model.fit_transform(images)
plt.scatter(umap_fit[:, 0], umap_fit[:, 1], c=labels, s=10)
plt.show()
```
